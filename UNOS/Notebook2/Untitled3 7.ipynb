```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    'Race': ['White', 'Black', 'Hispanic', 'White', 'Asian', 'Black', 'Hispanic', 'White', 'Black'],
    'TypeOfDeath': ['Accident', 'Natural', 'Homicide', 'Suicide', 'Accident', 'Natural', 'Homicide', 'Accident', 'Suicide'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Female']
}
df = pd.DataFrame(data)

# Count occurrences
race_death_counts = df.groupby(['Race', 'TypeOfDeath', 'Gender']).size().reset_index(name='Count')

# FacetGrid for Gender
g = sns.catplot(
    data=race_death_counts, 
    x='Race', y='Count', hue='TypeOfDeath', 
    col='Gender', kind='bar', height=5, aspect=1.2, palette='Set2'
)
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Relationship Between Race, Type of Death, and Gender')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame creation (replace this with your actual DataFrame)
data = {
    'HemodynamicsRegistration_SYS_CAN': [120, np.nan, 130, 125],
    'HemodynamicsRegistration_PA_DIA_CAN': [15, 14, np.nan, 16],
    'HemodynamicsRegistration_PCW_CAN': [8, 7, 9, np.nan],
    'HemodynamicsRegistration_PA_MN_CAN': [10, 11, np.nan, 12]
}

df = pd.DataFrame(data)

# Fill NaN values with the median of each column
df['HemodynamicsRegistration_SYS_CAN'] = df['HemodynamicsRegistration_SYS_CAN'].fillna(df['HemodynamicsRegistration_SYS_CAN'].median())
df['HemodynamicsRegistration_PA_DIA_CAN'] = df['HemodynamicsRegistration_PA_DIA_CAN'].fillna(df['HemodynamicsRegistration_PA_DIA_CAN'].median())
df['HemodynamicsRegistration_PCW_CAN'] = df['HemodynamicsRegistration_PCW_CAN'].fillna(df['HemodynamicsRegistration_PCW_CAN'].median())
df['HemodynamicsRegistration_PA_MN_CAN'] = df['HemodynamicsRegistration_PA_MN_CAN'].fillna(df['HemodynamicsRegistration_PA_MN_CAN'].median())

# Calculate the median for each row
row_medians = df.mean(axis=1)

# Add the row medians as a new column in the DataFrame
df['Row_Median'] = row_medians

# Display the updated DataFrame
print("DataFrame with Row Medians:")
```
