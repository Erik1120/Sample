```python
def write_csv(df, filename):
    """
    write dataframe to disk
    """
    # intialize variable
    file_path = '../Data/' + filename
    
    # write to disk
    df.to_csv(file_path, index=False)
    
    return print(f"{len(df):,} records written to {file_path}.")
```

```python
# write out csv file
filename = 'thoraticColumnNames.csv'
write_csv(thoratic_star, filename)

# flat file
filename = 'flatfile.csv'
write_csv(flat_star, filename)
```
